```python
from sympy import *
from sympy.plotting import plot3d
```

$𝒇(𝒙)=𝒙^𝟐+y^2$

```python
x,y = symbols('x y')
f = x**2 + y**2
plot3d(f,size=(10,7))
plt.show()
```

$𝒇(𝒙)=𝒙^𝟐+xy+y^2$  

```python
x,y = symbols('x y')
f = x**2 + x * y + y**2
plot3d(f,size=(10,7))
plt.show()
```

$𝒇(𝒙)=𝒙^𝟐y$ 

```python
x,y = symbols('x y')
f = x**2 * y
plot3d(f,size=(10,7))
plt.show()
```
